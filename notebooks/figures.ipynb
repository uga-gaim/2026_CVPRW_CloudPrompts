```python
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from pathlib import Path
from IPython.display import display

ROOT = Path("../inference/clipseg_cloudsen12plus_prompt_variants").expanduser().resolve()

def get_agg_row(df: pd.DataFrame, csv_path: Path) -> pd.Series:
    if "id" not in df.columns:
        raise ValueError(f"'id' column missing in {csv_path}")
    agg = df[df["id"] == "__aggregate__"]
    if agg.empty:
        raise ValueError(f"No __aggregate__ row found in {csv_path}")
    return agg.iloc[0]

iou_csvs = sorted(ROOT.glob("V*_iou.csv"))
if not iou_csvs:
    raise FileNotFoundError(f"No V*_iou.csv files found in: {ROOT}")

rows = []
for p in iou_csvs:
    tag = p.stem.replace("_iou", "")
    df = pd.read_csv(p)
    agg = get_agg_row(df, p)

    out = {"variant": tag}

    iou_cols = [c for c in df.columns if c.startswith("iou_")]
    if "miou" in df.columns:
        out["miou"] = float(agg["miou"])
    elif iou_cols:
        out["miou"] = float(pd.to_numeric(agg[iou_cols], errors="coerce").mean())
    else:
        out["miou"] = np.nan

    for c in iou_cols:
        out[c] = float(pd.to_numeric(agg[c], errors="coerce"))

    rows.append(out)

variant_df = pd.DataFrame(rows)

variant_df["vnum"] = variant_df["variant"].str.extract(r"V(\d+)", expand=False).astype(int)
variant_df = variant_df.sort_values("vnum").drop(columns=["vnum"]).reset_index(drop=True)

display(variant_df)
```

```python
from pathlib import Path
import pandas as pd

ZS_MIOU_CSV = Path("../inference/clipseg_cloudsen12plus_zs/miou_clipseg_cloudsen12plus_zs_eval.csv").expanduser().resolve()
print("ZS_MIOU_CSV:", ZS_MIOU_CSV, "exists:", ZS_MIOU_CSV.exists())

zs_df = pd.read_csv(ZS_MIOU_CSV)
zs_agg = get_agg_row(zs_df, ZS_MIOU_CSV)

zs_vals = {
    "miou": float(zs_agg["miou"]),
    "iou_clear": float(zs_agg["iou_clear"]),
    "iou_thick_cloud": float(zs_agg["iou_thick_cloud"]),
    "iou_thin_cloud": float(zs_agg["iou_thin_cloud"]),
    "iou_cloud_shadow": float(zs_agg["iou_cloud_shadow"]),
}

zs_vals
```

```python
def plot_variants_vs_zs(metric: str, ylabel: str, ylim=None):
    x_labels = variant_df["variant"].tolist()
    y = variant_df[metric].astype(float).tolist()
    z = zs_vals[metric]

    plt.figure(figsize=(10, 5))

    plt.plot(
        range(len(x_labels)),
        y,
        marker="o",
        linestyle="-",
        color="#f71735",
        label="Prompt Variants"
    )

    plt.axhline(
        z,
        linestyle="--",
        color="#044389",
        label="Zero-shot baseline"
    )

    plt.xticks(range(len(x_labels)), x_labels, fontsize=13, rotation=0)
    plt.yticks(fontsize=15)
    plt.ylabel(ylabel, fontsize=20)
    if ylim is not None:
        plt.ylim(ylim)
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.legend(fontsize=14)
    plt.tight_layout()
    plt.show()

plot_variants_vs_zs("miou", "mIoU", ylim=(0.0, 0.275))
```

```python
CM_ZS_PATH = "../inference/clipseg_cloudsen12plus_zs/cm_clipseg_cloudsen12plus_zs_eval.csv"
CM_LORA_PATH = "../inference/clipseg_cloudsen12plus_lora/cm_clipseg_cloudsen12plus_lora_eval.csv"
CM_FT_PATH = "../inference/clipseg_cloudsen12plus_fft/cm_clipseg_cloudsen12plus_fft_eval.csv"

CLASS_NAMES = ["Clear", "Thick Cloud", "Thin Cloud", "Cloud Shadow"]
N_CLASSES = len(CLASS_NAMES)
```

```python
cm_zs = pd.read_csv(CM_ZS_PATH)
cm_lora = pd.read_csv(CM_LORA_PATH)
cm_ft = pd.read_csv(CM_FT_PATH)
```

```python
ids_zs = set(cm_zs["id"]) - {"__aggregate__"}
ids_l = set(cm_lora["id"]) - {"__aggregate__"}
ids_f = set(cm_ft["id"]) - {"__aggregate__"}

common_ids = sorted(list(ids_zs & ids_l & ids_f))
```

```python
cm_zs_p = cm_zs[cm_zs["id"].isin(common_ids)].copy()
cm_lora_p = cm_lora[cm_lora["id"].isin(common_ids)].copy()
cm_ft_p = cm_ft[cm_ft["id"].isin(common_ids)].copy()

def confusion_from_row(row, n=N_CLASSES, prefix="cm_"):
    m = np.zeros((n, n), dtype=np.int64)
    for i in range(n):
        for j in range(n):
            col = f"{prefix}t{i}_p{j}"
            m[i, j] = int(row[col])
    return m

def aggregate_confusion(df, n=N_CLASSES):
    mats = []
    for _, r in df.iterrows():
        mats.append(confusion_from_row(r, n=n))
    return np.sum(mats, axis=0)

CM_ZS_GLOBAL = aggregate_confusion(cm_zs_p)
CM_LORA_GLOBAL = aggregate_confusion(cm_lora_p)
CM_FT_GLOBAL = aggregate_confusion(cm_ft_p)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def _normalize_cm(cm, normalize="true"):
    cm = cm.astype(float)
    if normalize == "true":
        denom = cm.sum(axis=1, keepdims=True)
    elif normalize == "pred":
        denom = cm.sum(axis=0, keepdims=True)
    else:
        denom = 1.0
    return np.divide(cm, denom, out=np.zeros_like(cm, dtype=float), where=denom != 0)


def plot_confusions_grid(
    cms,
    titles,
    normalize="true",
    cmaps=None,
    figsize=(18, 5),
    vmin=0.0,
    vmax=1.0,
    annot=True,
    annot_fmt=".2f",
    annot_size=9,
    annot_colors=None,
    colorbar_label="",
    title_size=12,
    tick_size=9,
    xrot=0,
    yrot=90,
    show_single_colorbar=True,
    show_shared_axis_labels=True,
):
    n = len(cms)
    if cmaps is None:
        cmaps = ["Blues"] * n
    if annot_colors is None:
        annot_colors = ["black"] * n

    fig, axes = plt.subplots(1, n, figsize=figsize, constrained_layout=True, sharex=True, sharey=True)
    if n == 1:
        axes = [axes]

    last_im = None

    for idx, (ax, cm, title) in enumerate(zip(axes, cms, titles)):
        cmn = _normalize_cm(cm, normalize=normalize)

        last_im = ax.imshow(
            cmn,
            cmap=cmaps[idx],
            vmin=vmin,
            vmax=vmax,
            interpolation="nearest"
        )

        ax.set_xticks(np.arange(N_CLASSES))
        ax.set_yticks(np.arange(N_CLASSES))


        ax.annotate(
            title,
            xy=(0.5, -0.14),
            xycoords="axes fraction",
            ha="center",
            va="bottom",
            fontsize=title_size,
            annotation_clip=False
        )

        ax.set_xticklabels(CLASS_NAMES, rotation=xrot, ha="center", fontsize=tick_size)
        ax.tick_params(axis="x", pad=4)

        if idx == 0:
            ax.set_yticklabels(CLASS_NAMES, rotation=yrot, ha="right", va="center", fontsize=tick_size)
            ax.tick_params(axis="y", pad=4)
        else:
            ax.tick_params(axis="y", left=False, labelleft=False)

        if not show_shared_axis_labels:
            ax.set_xlabel("Predicted", fontsize=tick_size + 1)
            if idx == 0:
                ax.set_ylabel("Actual", fontsize=tick_size + 1)

        if annot:
            for i in range(N_CLASSES):
                for j in range(N_CLASSES):
                    ax.text(
                        j, i, format(cmn[i, j], annot_fmt),
                        ha="center", va="center",
                        fontsize=annot_size,
                        color=annot_colors[idx]
                    )

    if show_shared_axis_labels:
        fig.supxlabel("Prediction", fontsize=tick_size + 8, y=-0.06)
        fig.supylabel("Ground Truth", fontsize=tick_size + 8)

    if show_single_colorbar and last_im is not None:
        cbar = fig.colorbar(last_im, ax=axes, location="right", fraction=0.025, pad=0.02)
        cbar.set_label(
            "Row-normalized Proportion",
            rotation=90,
            labelpad=18,
            fontsize=tick_size + 2
        )
        cbar.ax.yaxis.set_label_position("left")
        cbar.ax.tick_params(labelsize=tick_size)

    plt.show()


plot_confusions_grid(
    cms=[CM_ZS_GLOBAL, CM_LORA_GLOBAL, CM_FT_GLOBAL],
    titles=["(a) Zero-shot", "(B) LoRA", "(C) FFT"],
    normalize="true",
    cmaps=["YlOrBr", "YlOrBr", "YlOrBr"],
    figsize=(20, 6),
    vmin=0, vmax=1,
    annot=True,
    annot_size=12,
    annot_colors=["black", "black", "black"],
    title_size=18,
    tick_size=12,
    xrot=0,
    yrot=0,
    show_single_colorbar=True,
    show_shared_axis_labels=True,
)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

ZS_CSV   = "../inference/clipseg_cloudsen12plus_zs/miou_clipseg_cloudsen12plus_zs_eval.csv"
LORA_CSV = "../inference/clipseg_cloudsen12plus_lora_low_data/lora_low_data_ladder.csv"
FFT_CSV  = "../inference/clipseg_cloudsen12plus_fft_low_data/fft_low_data_ladder.csv"

LORA_FULL_CSV = "../inference/clipseg_cloudsen12plus_lora/miou_clipseg_cloudsen12plus_lora_eval.csv"
FFT_FULL_CSV  = "../inference/clipseg_cloudsen12plus_fft/miou_clipseg_cloudsen12plus_fft_eval.csv"

zs_df   = pd.read_csv(ZS_CSV)
lora_df = pd.read_csv(LORA_CSV)
fft_df  = pd.read_csv(FFT_CSV)
lora_full_df = pd.read_csv(LORA_FULL_CSV)
fft_full_df  = pd.read_csv(FFT_FULL_CSV)

lora_df = lora_df.sort_values("pct").reset_index(drop=True)
fft_df  = fft_df.sort_values("pct").reset_index(drop=True)

zs_agg = zs_df[zs_df["id"] == "__aggregate__"]
if zs_agg.empty:
    raise ValueError("No __aggregate__ row found in zero-shot CSV.")
zs_agg = zs_agg.iloc[0]



def get_aggregate_row(df, name):
    agg = df[df["id"] == "__aggregate__"]
    if agg.empty:
        raise ValueError(f"No __aggregate__ row found in {name}.")
    return agg.iloc[0]


lora_full_agg = get_aggregate_row(lora_full_df, "LORA_FULL_CSV")
fft_full_agg  = get_aggregate_row(fft_full_df, "FFT_FULL_CSV")


lora_100_row = {col: np.nan for col in lora_df.columns}
lora_100_row["pct"] = 100.0
if "miou_mean" in lora_100_row:
    lora_100_row["miou_mean"] = float(lora_full_agg["miou"])
if "iou_clear_mean" in lora_100_row and "iou_clear" in lora_full_agg.index:
    lora_100_row["iou_clear_mean"] = float(lora_full_agg["iou_clear"])
if "iou_thick_cloud_mean" in lora_100_row and "iou_thick_cloud" in lora_full_agg.index:
    lora_100_row["iou_thick_cloud_mean"] = float(lora_full_agg["iou_thick_cloud"])
if "iou_thin_cloud_mean" in lora_100_row and "iou_thin_cloud" in lora_full_agg.index:
    lora_100_row["iou_thin_cloud_mean"] = float(lora_full_agg["iou_thin_cloud"])
if "iou_cloud_shadow_mean" in lora_100_row and "iou_cloud_shadow" in lora_full_agg.index:
    lora_100_row["iou_cloud_shadow_mean"] = float(lora_full_agg["iou_cloud_shadow"])


fft_100_row = {col: np.nan for col in fft_df.columns}
fft_100_row["pct"] = 100.0
if "miou_mean" in fft_100_row:
    fft_100_row["miou_mean"] = float(fft_full_agg["miou"])
if "iou_clear_mean" in fft_100_row and "iou_clear" in fft_full_agg.index:
    fft_100_row["iou_clear_mean"] = float(fft_full_agg["iou_clear"])
if "iou_thick_cloud_mean" in fft_100_row and "iou_thick_cloud" in fft_full_agg.index:
    fft_100_row["iou_thick_cloud_mean"] = float(fft_full_agg["iou_thick_cloud"])
if "iou_thin_cloud_mean" in fft_100_row and "iou_thin_cloud" in fft_full_agg.index:
    fft_100_row["iou_thin_cloud_mean"] = float(fft_full_agg["iou_thin_cloud"])
if "iou_cloud_shadow_mean" in fft_100_row and "iou_cloud_shadow" in fft_full_agg.index:
    fft_100_row["iou_cloud_shadow_mean"] = float(fft_full_agg["iou_cloud_shadow"])


lora_df = pd.concat([lora_df, pd.DataFrame([lora_100_row])], ignore_index=True)
fft_df  = pd.concat([fft_df,  pd.DataFrame([fft_100_row])],  ignore_index=True)


lora_df = lora_df.drop_duplicates(subset=["pct"], keep="last").sort_values("pct").reset_index(drop=True)
fft_df  = fft_df.drop_duplicates(subset=["pct"], keep="last").sort_values("pct").reset_index(drop=True)


METRICS = [
    ("mIoU (All classes)", "miou_mean", "miou"),
    ("Clear IoU", "iou_clear_mean", "iou_clear"),
    ("Thick Cloud IoU", "iou_thick_cloud_mean", "iou_thick_cloud"),
    ("Thin Cloud IoU", "iou_thin_cloud_mean", "iou_thin_cloud"),
    ("Cloud Shadow IoU", "iou_cloud_shadow_mean", "iou_cloud_shadow"),
]

print("Loaded:")
print("  ZS rows       :", len(zs_df))
print("  LoRA rows     :", len(lora_df), "(includes 100%)")
print("  FFT rows      :", len(fft_df),  "(includes 100%)")
print("\nZero-shot aggregate baseline values:")
for title, _, zs_col in METRICS:
    print(f"  {title:<20} = {zs_agg[zs_col]:.4f}")

print("\nAdded 100% points:")
print(f"  LoRA 100% mIoU = {lora_100_row['miou_mean']:.4f}")
print(f"  FFT  100% mIoU = {fft_100_row['miou_mean']:.4f}")
```

```python
KEEP_PCTS = None          
Y_LIM = (-7.5, 37.5)             
FIGSIZE = (10, 5)

IMPROV_COL = "miou_pct_improvement_from_prev"


lora_gain = lora_df.copy()
fft_gain  = fft_df.copy()

for df in (lora_gain, fft_gain):
    df["pct"] = pd.to_numeric(df["pct"], errors="coerce")
    df["miou_mean"] = pd.to_numeric(df["miou_mean"], errors="coerce")
    df.dropna(subset=["pct", "miou_mean"], inplace=True)
    df.sort_values("pct", inplace=True)
    df.reset_index(drop=True, inplace=True)

if KEEP_PCTS is not None:
    keep_set = set(float(x) for x in KEEP_PCTS)
    lora_gain = lora_gain[lora_gain["pct"].isin(keep_set)].reset_index(drop=True)
    fft_gain  = fft_gain[fft_gain["pct"].isin(keep_set)].reset_index(drop=True)


lora_gain[IMPROV_COL] = lora_gain["miou_mean"].pct_change() * 100.0
fft_gain[IMPROV_COL]  = fft_gain["miou_mean"].pct_change() * 100.0


zs_miou = float(zs_agg["miou"])


lora_mask_01 = np.isclose(lora_gain["pct"].astype(float), 0.1)
if lora_mask_01.any():
    lora_idx_01 = lora_gain.index[lora_mask_01][0]
    lora_gain.loc[lora_idx_01, IMPROV_COL] = (
        (lora_gain.loc[lora_idx_01, "miou_mean"] - zs_miou) / zs_miou
    ) * 100.0


fft_mask_01 = np.isclose(fft_gain["pct"].astype(float), 0.1)
if fft_mask_01.any():
    fft_idx_01 = fft_gain.index[fft_mask_01][0]
    fft_gain.loc[fft_idx_01, IMPROV_COL] = (
        (fft_gain.loc[fft_idx_01, "miou_mean"] - zs_miou) / zs_miou
    ) * 100.0


all_pcts = sorted(set(lora_gain["pct"].tolist()) | set(fft_gain["pct"].tolist()))
x_pos = np.arange(len(all_pcts))
pct_to_x = {pct: i for i, pct in enumerate(all_pcts)}
x_labels = [f"{v:g}" for v in all_pcts]


lora_plot = lora_gain.dropna(subset=[IMPROV_COL]).copy()
fft_plot  = fft_gain.dropna(subset=[IMPROV_COL]).copy()

lora_x = lora_plot["pct"].map(pct_to_x).to_numpy()
fft_x  = fft_plot["pct"].map(pct_to_x).to_numpy()



plt.figure(figsize=FIGSIZE)

plt.plot(
    fft_x, fft_plot[IMPROV_COL],
    marker="s", linestyle="-", color="#00916e", label="FFT"
)
plt.plot(
    lora_x, lora_plot[IMPROV_COL],
    marker="o", linestyle="-", color="#ffcf00", label="LoRA"
)

plt.axhline(0, linestyle="--", color="#044389", label="No change")

plt.xlabel("Training Data (%)", fontsize=20)
plt.ylabel("mIoU Improvement (%)", fontsize=20)

plt.xticks(x_pos, x_labels, fontsize=15)
plt.yticks(fontsize=15)
plt.xlim(-0.3, len(all_pcts) - 0.7)

if Y_LIM is not None:
    plt.ylim(Y_LIM)

plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=14)
plt.tight_layout()
plt.show()
```
