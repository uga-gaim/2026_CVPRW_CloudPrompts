```python
import datetime

# Check ASOS date range
print("ASOS data date range:")
print(f"  Start: {asos_df['timestamp'].min()}")
print(f"  End: {asos_df['timestamp'].max()}")

# Check a sample image timestamp
sample_image = '../data/7410/snap_7410_20250615_191911_6d6cd753.jpg'
try:
    sample_time = labeling.extract_timestamp_from_image(sample_image)
    print(f"\nSample image timestamp: {sample_time}")
    print(f"Prediction window would be: {sample_time + datetime.timedelta(minutes=60)} to {sample_time + datetime.timedelta(minutes=120)}")
except Exception as e:
    print(f"Error: {e}")
```

```python
import glob, os
image_dir = '../data/7410'
abs_dir = os.path.abspath(image_dir)
print("Expecting images in:", abs_dir)
print("Sample files:", glob.glob(abs_dir + '/*.jpg')[:5])
```

```python
for path in glob.glob(abs_dir + '/*.jpg')[:10]:
    try:
        ts = labeling.extract_timestamp_from_image(path)
        print(os.path.basename(path), "→", ts)
    except Exception as e:
        print("Failed to parse:", os.path.basename(path), e)
```

```python
import pandas as pd
asos_df['timestamp'] = pd.to_datetime(asos_df['timestamp']).dt.tz_localize(None)
```

```python
sample = glob.glob(abs_dir + '/*.jpg')[0]
img_dt = labeling.extract_timestamp_from_image(sample)
print("Image datetime:", img_dt)
print("Window \n", 
      asos_df[(asos_df['timestamp'] >= img_dt + datetime.timedelta(minutes=60)) &
              (asos_df['timestamp'] < img_dt + datetime.timedelta(minutes=120))])
```
